```python
#Ahora usemos apply con labmda
#vamos a pasar los años a meses en la informacion de los jugadores
#En este caso, si usamos apply no necisitamos poner el nombre de la funcion.
df_jugadores['age'].apply(lambda x:x*12)
```

```python
#Ahora usemos apply con labmda
#vmos a pasar los cm a mt en la informacion de los jugadores
#En este caso, si usamos apply no necisitamos poner el nombre de la funcion.
df_jugadores['height_cm'].apply(lambda x:x/100)
```

```python
#Tambien puedo pasar los nombres de los jugadores a mayucular
df_jugadores['long_name'].apply(lambda x:x.upper())
```
