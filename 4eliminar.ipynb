```python
#tambien podemos eliminar columnas y flas con diferentes metodos
import pandas as pd
import numpy as np
```

```python
#Traigo el archivo que voy a trabajar
df_jugadores=pd.read_csv("/content/drive/Shareddrives/GALIAS_COMERCIAL/PYTHON/players_20 (1).csv")

df_jugadores
```

```python
#ahora fijamos el index
df_jugadores=df_jugadores.set_index('short_name')
```

```python
#vamos a traer solo las columnas que necesito, y voy actualizar el dataframe
df_jugadores=df_jugadores[["player_url",  "long_name",  "age",  "dob",  "height_cm",  "weight_kg",  "nationality",  "club"]]
df_jugadores.head(3)
```

```python
#Vamos a eliminar registros usando el drop
#axis nos sirve para especificar si quiero hacer la accion sobre la columna o sobre la fila, 0 para la fila 1 para la columna
df_jugadores.drop('L. Messi', axis=0)
```

```python
#Vamos a eliminar teniendo en cue ta el index
#como es index se entiende que es toda la fila
df_jugadores.drop(index=['Neymar Jr'])
```

```python
#ahora vamos a eliminar dos o mas filas pero las vaos a actualizar
df_jugadores=df_jugadores.drop(['L. Messi','Neymar Jr'], axis=0)
df_jugadores
```

```python
#ahora vamos a eliminar dos o mas filas pero las vaos a actualizar
df_jugadores=df_jugadores.drop(['E. Hazard','K. De Bruyne'], axis=0)
df_jugadores
```

```python
#vamos a eliminar una columna, con axis 1 se elimina toda la colunma
df_jugadores.drop('player_url', axis=1)
```

```python
#otro metodo para eliminar colunmas sin usar el axis es el siguiente
df_jugadores.drop(columns=['nationality'])
```

```python
#para eliminar la ultima columna por posidion si no la supieramos
df_jugadores.drop(df_jugadores.columns[[-1]], axis=1)
```

```python
#para eliminar la ultima columna por posidion si no la supieramos
df_jugadores.drop(df_jugadores.columns[[6]], axis=1)
```

```python
#para eliminar mas de una columna
#si quiero reescribir la base la debo guardar en una variable
df_jugadores=df_jugadores.drop(['long_name', 'dob' ], axis=1)
```
