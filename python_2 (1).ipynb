```python
#Importro pandas y le pongo el sobre nombre pd y nunpy como np
import pandas as pd
import numpy as np
```

```python
#La primara forma de crear un dataframe es usando el metodo array, y luego guardarlo en una variable.
dataframe_1= np.array([[2,4],[6,8],[10,12],[14,16]])
```

```python
#Para crear el dataframe del arreglo anterior debemos unsar pandas.
pd.DataFrame(dataframe_1)
```

```python
pd.DataFrame(dataframe_2)
```
