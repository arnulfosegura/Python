```python
#vamos a usar el metodo pivotprimero
#Lo primero es importar librerias
import pandas as pd
import numpy as np
```

```python
#Traemos el archivo
df_poblacion=pd.read_csv('/content/drive/Shareddrives/GALIAS_COMERCIAL/PYTHON/population_total (1).csv', encoding='ISO-8859-1')
df_poblacion.head(3)
```

```python
#lo primero que vamos a hacer es eliminar los null o los vacios, con el metodod dropna
df_poblacion.dropna(inplace=True)
```

```python
#primero pivoteamos para que la visulizacion sea mas amigable
#recordemos que para pivotear tenemos 3 argumentos, el index, las columnas y valores
df_poblacion.pivot(index="year", columns="country", values="population")
```

```python
#creo un nuevo dataframe con la tabla que pivotee
df_pivot_indicadores=df_poblacion.pivot(index="year", columns="country", values="population")
df_pivot_indicadores[['Colombia', 'United States', 'India', 'China', 'Indonesia']]
```

```python
df_pivot_indicadores=df_pivot_indicadores[['Colombia', 'United States', 'India', 'China', 'Indonesia']]
```

```python
df_pivot_indicadores.index
```

```python
#identificamos que hay del 2020
df_pivot_indicadores.index.isin([2020])
```

```python
#llamamos lo que hay del 2020
df_pivot_indicadores[df_pivot_indicadores.index.isin([2020])]
```

```python
#Los voy a guardar en otro dataframe
df_pivot_indicadores_2020=df_pivot_indicadores[df_pivot_indicadores.index.isin([2020])]
```

```python
df_pivot_indicadores_2020
```

```python
#para transponer uso el metodo T
df_pivot_indicadores_2020.T
```
